```python
testProblem = {
    "parts": [1,2,3,4],
    "mark": {"shape": "circle", "color": "blue"},
    "piecePoss": {
        "blue":{"x":3, "y":14},
        "yellow":{"x":0, "y":0},
        "black":{"x":15, "y":14},
        "green":{"x":15, "y":12},
        "red":{"x":0, "y":14}
    },
    "torus":False
}
```

```python
import asyncio

loop = asyncio.get_event_loop()
for i in range(10):
    def task():
        rq.post("http://localhost:8080/solve", headers={"content-type":"application/json"}, data=json.dumps(
            {"problem":testProblem}
        ))
    loop.run_in_executor(None, task)
```

```python
import subprocess
import time
start_time = time.perf_counter()
sum = 0
n = 100
mx = 0
mn = 10000

for i in range(n):
    prob = f"api/app/solver/test/{i}"
    
    start_time = time.perf_counter()
    subprocess.call("api/app/solver/solver_g < "+prob, shell=True)
    end_time = time.perf_counter()

    elapsed_time = end_time - start_time
    mx = max(mx, elapsed_time)
    mn = min(mn, elapsed_time)
    print(elapsed_time)
    sum += elapsed_time

print("avg: ", sum / n)
print("max: ", mx)
print("mn: ", mn)
```
