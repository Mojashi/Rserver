```python
import subprocess
import time
start_time = time.perf_counter()
sum = 0
n = 100
mx = 0
mn = 10

for i in range(n):
    prob = f"app/solver/test/{i}"
    
    start_time = time.perf_counter()
    subprocess.call("app/solver/solver_g < "+prob, shell=True)
    end_time = time.perf_counter()

    elapsed_time = end_time - start_time
    mx = max(mx, elapsed_time)
    mn = min(mn, elapsed_time)
    print(elapsed_time)
    sum += elapsed_time

print("avg: ", sum / n)
print("max: ", mx)
print("mn: ", mn)
```
